## Data Visualization Activity

### Importing the Packages and Data

```python
import pandas as pd
import seaborn
from matplotlib import pyplot
```

#### Above are the packages used. Now, let's customize the visuals a bit more.

```python
%config InlineBackend.figure_format = 'retina'
seaborn.set_style(style='whitegrid')
```

```python
parkinsons = pd.read_csv('parkinsons.csv')
parkinsons.columns
```

#### The columns function was used to check the headers of each column, making sure the CSV loaded into the virtual environment correctly.

### Relational Plots

```python
seaborn.scatterplot(
    data=parkinsons,
    x='NHR',
    y='HNR'
);
```

#### It appears as though the relation between NHR and HNR is one of exponential decay.

```python
seaborn.scatterplot(
    data=parkinsons,
    x='NHR',
    y='HNR',
    hue='PPE',
    style='PPE'
).set(
    title='NHR and HRN',
    xlabel='NHR',
    ylabel='HRN'
);
```

#### After running this code, I realized there weren't many practical variables to use in place of style and hue. The figure legend is nightmarish.

```python
seaborn.scatterplot(
    data=parkinsons,
    x='NHR',
    y='HNR',
    hue='status',
    style='status'
).set(
    title='NHR and HRN',
    xlabel='NHR',
    ylabel='HRN'
);
```

#### This looks better. The status variable refers to if a patient has Parkinson's or not.

### Distribution Plots

```python
seaborn.displot(
    parkinsons,
    x='PPE'
).set(
    title='PPE distribution',
    xlabel='PPE',
    ylabel='Frequency'
);
```

#### It appears as though the PPE variable is left-skewed with a median somewhere around 0.2.

```python
seaborn.displot(
    parkinsons,
    x='PPE',
    hue='status'
).set(
    title='PPE distribution',
    xlabel='PPE',
    ylabel='Frequency'
)

pyplot.legend(
    title='Status'
);
```

```python
seaborn.displot(
    parkinsons,
    x='PPE',
    col='status'
);
```

#### Dividing the histogram by status shows a clearer picture that those with Parkinson's have higher PPE's than those who do not.

```python
seaborn.jointplot(
    parkinsons,
    x='NHR',
    y='HNR'
);
```

### Categorical Plots

```python
seaborn.catplot(
    data=parkinsons,
    x='status',
    y='PPE',
    kind='box'
).set(
    title='Status and PPE',
    xlabel='Status',
    ylabel='PPE'
);
```

#### Let's make this a bit more pleasing to the eye.

```python
seaborn.catplot(
    data=parkinsons,
    x='status',
    y='PPE',
    hue='status',
    kind='box'
).set(
    title='Status and PPE',
    xlabel='Status',
    ylabel='PPE'
);
```

```python
seaborn.catplot(
    data=parkinsons,
    x='status',
    y='PPE',
    hue='status',
    kind='violin'
).set(
    title='Status and PPE',
    xlabel='Status',
    ylabel='PPE'
);
```
